```python
def func(row):
    if row['gstm1_gene']=='Present':
        return 1
    elif row['gstm1_gene']=='Absent':
        return 0
    
df['gstm1_gene_nrisk']=df.apply(func, axis=1)

def func(row):
    if row['gstt1_gene']=='Present':
        return 1
    elif row['gstt1_gene']=='Absent':
        return 0
    
df['gstt1_gene_nrisk']=df.apply(func, axis=1)

df.iloc[:,4:8].head()
```

```python
rs_list=[]
for column in df.columns.tolist():
    if 'rs' in column:
        rs_list.append(column)
print(f'length: {len(rs_list)}')
rs_list
```

```python
lisst_unk = []

for column in df.columns.tolist():
    x= f'{column}: {df[column].unique()}'
    if 'Unknown' in x:
        lisst_unk.append(column)
        
lisst_unk[:10]
```

```python
for column in lisst_unk:
    x= f'{column}: {df[column].unique()}'
    if 'Unknown' in x:
        df[column] = df[column].replace("Unknown", np.NaN)
for column in df.columns.tolist():
    print(f'{column}: {df[column].unique()}')
```
